```python
l=[{'name':'pen' , 'price':23,'description':'Bluepoints','availability':100}]
def continue_task():
    c=input("Press 'a' to  continue and 'b' to end :")
    if(c=="a"):
        main()
    else:
        print("Thank you")
        
def add_product(a,b,c,d):
    products={}
    products['name']=a
    products['price']=b
    products['description']=c
    products['availability']=d
    
    return products
    
def add():
    a=input("Product name: ")
    b=float(input("Price: "))
    c=input("Description: ")
    d=int(input("Availability: "))
    v=add_product(a,b,c,d)
    l.append(v)
    print("Product added successfully")


def show_products():
    for i in l:
        print(i)

def update_product():
    a=input("Enter the product name to update: ")
    for i in l:
        d=0
        if a in i.values():
            b,c=input("Enter the key and value to be updated: ").split()
            i[b]=c
            break
        else:
            d=1
            pass
    if d == 1:
        print(f'Product {a} is not available')
        
def stock():
    a=input("Enter the product name to know its stock level: ")
    for i in l:
        d=0
        if a in i.values():
            print(i['availability'])
            break
        else:
            d=1
            pass
    if d == 1:
        print(f'Product {a} is not available')

def update_stock(a,b,c):
    for i in l:
        if a in i.values() and i['name']==a and c==1:
            i['availability']=int(i['availability'])-b
        else:
            i['availability']=int(i['availability'])+b
            
    

def place_orders():
    a=input("Enter the product you need to purchase: ")
    b=int(input("Enter the number of product you need:  "))
    for i in l:
        d=0
        if a in i.values():
            print(f'Orders placed : {a} total order {b} ')
            if(b>i['availability'] and i['name']==a):
                print("Out of stock : Quantity Availabilty :"+str(i['availability']))
            else:
                update_stock(a,b,c=1)
            break
        else:
            d=1
            pass
    if d == 1:
        print(f'Product {a} is not available')
        
def cancel_orders():
    a=input("Enter the product ordered that need to cancel: ")
    b=int(input("Enter the number of product you bought:  "))
    for i in l:
        d=0
        if a in i.values():
            print(f'Order cancelled: {a} total order {b} ')
            update_stock(a,b,c=2)
            break
        else:
            d=1
            pass
    if d == 1:
        print(f'Product {a} is not available')

def main():
    print('''E-commerce Inventory Management System
    Provide the task to be implemented:
    1.Product management
    2.Stock management
    3.Order Processing
    4.Inventory reports''')
    a=int(input("Enter option val: "))
    if(a==1):
        print('''Product management:
        1.Add product
        2.Update product''')
        b=int(input("Provide the value: "))
        if b==1:
            add()
            continue_task()
        elif b==2:
            update_product()
            continue_task()
    elif(a==2):
        print("Stock management: ")
        stock()
        continue_task()
        
    elif(a==3):
        print('''Order Processing: 
        1.Place order 
        2.Cancel order''')
        b=int(input("Provide the task val: "))
        if b == 1:
            place_orders()
            continue_task()
        elif b==2 :
            cancel_orders()
            continue_task()
    elif(a==4):
        print("Inventory Reports of all products")
        show_products()
        continue_task()
            
            
            
if __name__=="__main__":
    main()
```
