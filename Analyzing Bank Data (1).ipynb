```python
'''The bank needs to know the number of male
customers who are:
1. "Graduate"
2. Have the loan status as 'Y (yes)â€˜
'''
bank[(bank['Education']=='Graduate') & (bank['Loan_Status']=='Y')]
```

```python
'''â€¢ To decide on loan approval for female
customers, the bank officer wishes to see past
records of female customers who:
1. Are 'Self Employed'
2. Have an income greater than $4.000
â€¢ Help the bank find and display the records of
such customers.'''

bank[(bank['Gender']=='Female') & (bank['Self_Employed']=='Yes') & (bank['ApplicantIncome'] > 4000)]
```

```python
'''The bank officer wishes to see the history of
loan approval of customers to ensure they
have:
1. More than three dependents,
2. Have 'Rural property area,
3. Have the loan status as 'Approved'
â€¢ Help the bank officer accomplish this task and
determine the number of such customers'''

#bank.Dependents.replace('3+',bank.Dependents.mode,inplace='True')

count1=bank[(bank.Dependents > '3') & (bank.Property_Area=='Rural') & (bank.Loan_Status=='Y')];
count1.Dependents.count()
```

```python
'''The customer Wamer K' has applied for a bank
loan of USD 290.000. The bank officer needs
to know the average income of the previous
customers for whom a loan worth $290,000 or
more was approved.
â€¢ Help the bank officer accomplish this task.'''

bank[(bank.LoanAmount >= 290) & (bank.Loan_Status=='Y')]['ApplicantIncome'].mean
```

```python
'''The bank officer wishes to know why the loan
got rejected for the customers who:
1. Are unmarried
2. self-employed
â€¢ Help the bank officer display the record of
such customers. Also, find out how many male
and female customers fall into this category.'''
bank[(bank.Married=='No') & (bank.Self_Employed=='Yes') & (bank.Loan_Status=='N')]
```

```python
bank[(bank.Married=='No') & (bank.Self_Employed=='Yes') & (bank.Loan_Status=='N')].groupby('Gender')['Loan_ID'].count()
```

```python
'''The applicant's income and property area play
an important role in the approval of a loan
application. The bank wishes to find out the
applicant's average income by area. How can
this task be accomplished?'''

df1=bank.groupby('Property_Area')['ApplicantIncome'].mean();
df1=pd.DataFrame(df1);df1
```

```python
'''â€¢ The bank officer needs to collect information
about married customers.
â€¢ Determine the numbers of customers whose
co-applicant's income is less than the average
of the incomes of all co-applicants and have
gotten their loan approved.'''


bank[(bank.CoapplicantIncome < bank.CoapplicantIncome.mean()) & (bank.Loan_Status=='Y') & (bank.Married=='Yes')].count()
```

```python
'''â€¢ As per the previous record of customers who
have applied for a bank loan, the bank officer
needs to know the number of customers who:
1. Are graduates
2. Have an income ranging between 10,256
and150,000.
â€¢ Help the bank officer accomplish this task.'''

bank[(bank['Education']=='Graduate') & (bank.ApplicantIncome.between(10256,150000))]
```
