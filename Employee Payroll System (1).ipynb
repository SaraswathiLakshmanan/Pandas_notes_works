```python
    
def payroll():
    for i in list_employees:
        print(str(i['Name'])+"-----"+str(i['Emp_id'])+"-----"+str(i['Position']))
        print("Payroll Detail:Basic Sal :{}\nHRA :{}\nDA :{}\nTax :{}\nNet_pay :{}".format(i['Salary'],i['HRA'],i['DA'],i['Tax'],i['Net_pay']))
def add_employee(a,b,c,d):
    employee_records={}
    employee_records['Name']=a
    employee_records['Emp_id']=b
    employee_records['Position']=c
    employee_records['Salary']=d
    list_employees.append(employee_records)
    print("New Record added to Employee Table ")


def get_detail():
    l=[]
    name=input("Enter Employee name: ")
    position=input("Enter position of the employee: ")
    sal=float(input("Enter the salary: "))
    l=[name,position,sal]
    return l

def show_employee():
    for i in list_employees:
        print(i)

def del_employee():
    a=employee_check()
    if(employee_check):
        for i in list_employees:
            if i['Name']==a:
                list_employees.remove(i)
                print("Deleted Successfully")
                break
        else:
            print("Employee Name is not in list")

def update_employee():
    a=employee_check()
    if(employee_check):
        b=input("Enter column to be updated:\n1.Name \n2.Position \n3.Emp_id \n4.Salary\n")
        c=input("Enter the value: ")
        for i in list_employees:
            if i['Name'] == a:
                if(b=='Name' or b=='Position'):
                    i[b]=c
                elif(b=='Emp_id'):
                    i[b]=int(c)
                else:
                    i[b]=float(c)
                break
                print("Employee detail Updated ")

def employee_check():
    a=input("Enter Employee name: ")
    count=False
    for i in list_employees:
        if i['Name']==a:
            count=True
            return a
    else:
        print("Employee not in list")

def cal_net_pay():
    a=employee_check()
    if(employee_check):
        for i in list_employees:
            if i['Name']==a:
                if(i['Salary']<10000):
                    hra=i['Salary']*0.05
                    da=0
                    tax=0
                    net_pay=i['Salary']+hra
                elif(i['Salary']>=10000 and i['Salary']<20000):
                    hra=i['Salary']*0.075
                    da=i['Salary']*0.05
                    tax=i['Salary']*0.05
                    net_pay=i['Salary']+hra+da-tax
                elif(i['Salary']>=20000):
                    hra=i['Salary']*0.1
                    da=i['Salary']*0.1
                    tax=i['Salary']*0.15
                    net_pay=i['Salary']+hra+da-tax
                i['Net_pay']=net_pay
                i['HRA']=hra
                i['DA']=da
                i['Tax']=tax
                print("Net_pay : {}".format(net_pay))
        
def main():
    global emp_id
    c=input("Press a to continue or b to end: ")
    if(c=='a'):
        print("1.Add employee\n2.Show employees\n3.Delete employee\n4.Update employee\n5.Calculate Net_Pay\n6.Payroll Details\n")
        d=int(input("Enter choice: "))
        if(d==1):
            c=get_detail()
            add_employee(c[0],emp_id,c[1],c[2])
            emp_id+=1
            main()
        elif(d==2):
            show_employee()
            main()
        elif(d==3):
            del_employee()
            main()
        elif(d==4):
            update_employee()
            main()
        elif(d==5):
            cal_net_pay()
            main()
        elif(d==6):
            payroll()
            main()
    else:
        print("Thank you")
        
if __name__=="__main__":
    print("Welcome to Employee Page")
    list_employees=[]
    emp_id=100
    main()
            
    
```
